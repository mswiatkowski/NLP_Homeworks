```python
# PUNKT 6: usuwamy nie-litery i słowa krótsze niż 2 znaki

def reject_non_grata(dictionary, length=2):
    list_of_non_grata = "<>,.:\";'?/{}[]|\+=_-)(*&^%$#@!~`)'1234567890 \n"
    new_dictionary = dict()
    for key in dictionary.keys():
#         print(key)
        if len(key) >= length and not any(letter for letter in list(key) if letter in list_of_non_grata): # sprawdzenie długości i obecności nie-liter
#             print(len(key))
            new_dictionary[key] = dictionary[key]
    return new_dictionary
```

```python
def sort_dict(dictionary):
    # mała gimnastyka w celu posortowania słownika:
    keys = list(dictionary.keys())
    keysInt = []
    for key in keys:
        keysInt.append(int(key))
    keysInt.sort()
    keys = []
    for key in keysInt:
        keys.append(str(key))
```

```python
def make_ranks(dictionary, list_of_ranks=10):
#     print(max(dictionary.values()))
    top = max(dictionary.values()) + 100
    tenth = int(top/(list_of_ranks))
#     print(top)
    buckets = [item for item in range(0, top, tenth)]
#     print(buckets)
    return buckets
```

```python
def assign_to_ranks(dictionary, list_of_ranks):
    ranks = dict()
    i = 1
    for elem in list_of_ranks:
        ranks[i] = elem
        i += 1
    word_ranked = dict()
    for item in dictionary.items():
#         print(item)
        list_of_mins = []
        for rank_val in ranks.values():
            list_of_mins.append((item[1]-rank_val)**2)
#         print(list_of_mins)
        minimum = min(list_of_mins)
#         print('minimum: ',minimum)
        min_index = list_of_mins.index(minimum)
        for key in ranks.keys():
            if key == min_index+1:
                if key not in word_ranked.keys():
                    word_ranked[key] = []
                word_ranked[key].append(item[0])
#     print(ranks)
    return word_ranked
```

```python
dict_test = {"bobo": 54233, "kaka": 564, "kiki": 546}
buckets = make_ranks(dict_test)
print(buckets)
assign_to_ranks(dict_test, buckets)
```
